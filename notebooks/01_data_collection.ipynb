### Measurement Method/測定方法

Physical characteristics of grape products were measured using two different methods.

Grain diameter was estimated using grape images that included a 100-yen coin as a scale reference.  
The coin served as a visual benchmark to approximate the size of individual grape grains.

Bunch dimensions, including bunch length and bunch width, were measured directly using a flexible tape measure.

These measurements were recorded as numerical variables in the dataset.

ぶどうの外観特性は、2つの方法によって測定しました。

粒径については、100円硬貨をスケールの基準として含めたぶどう画像を用い、
その大きさを目安に粒の直径を推定しました。

一方、房の長さおよび房の幅については、
一般的なメジャー（軟尺）を用いて直接測定しました。

これらの測定値は、データセット内の数値変数として記録されています。

### Variety System Classification/品種系統の分類

In this analysis, grape varieties were grouped into two broad variety systems:
**Shine Muscat-type** and **Kyoho-type**.

The classification was not based on formal genetic lineage.
Instead, varieties were grouped according to commonly recognized
market and sensory characteristics.

In general:

- **Shine Muscat-type varieties** tend to have a lighter, refreshing sweetness
  and muscat-like aroma.
- **Kyoho-type varieties** are typically characterized by a richer,
  juicier flavor profile.

This classification approach reflects how consumers often perceive
and differentiate grape varieties in retail markets.


本分析では、ぶどう品種を **「シャインマスカット系」** と
**「巨峰系」** の2つの大きな系統に分類しました。

この分類は、正式な遺伝系統に基づくものではありません。
代わりに、市場で一般的に認識されている
**味わいや香りの特徴**に基づいて分類を行いました。

一般的に、

- **シャインマスカット系**は、マスカット香を伴う
  さっぱりとした甘さが特徴です。
- **巨峰系**は、より濃厚でジューシーな味わいを持つ
  品種が多いとされています。

この分類は、消費者が市場でぶどうを選ぶ際の
感覚的な違いを反映することを目的としています。

```python
# List of varieties classified as Shine Muscat-type (Green Group)
GREEN = "Shine Muscat-type"
PURPLE = "Kyoho-type"

green_list = [
    'Queen Seven', 'Haneou', 'Queen Rouge', 'Shine Muscat',
    'Fujinokagayaki', 'Seto Giants', 'Shigyoku', 'Shinku', 'Kaiji',
    'Queen Roug', 'Scarlet', 'Red Shine Muscat', 'Suiho',
    'Sunshine Red', 'Ving', 'Wasekaiji', 'Pizzutello Bianco',
    'Nouvelle Rose', 'Muscat Noir', 'Miwahime'
]

def grape_group(variety):
    if variety in green_list:
        return GREEN
    else:
        return PURPLE

# Apply classification
df["group"] = df["grape_variety"].apply(grape_group)
```

Varieties not included in the predefined list were treated as
Kyoho-type varieties.

事前に定義したリストに含まれていない品種は、
巨峰系として扱いました。
