### Data Source

The dataset was collected manually from Japanese retail markets.
Each observation corresponds to a retail grape product with recorded
variety information, physical characteristics, and price.

### Variable Definition

- `origin_prefecture`: prefecture of origin
- `grape_variety`: commercial grape variety name
- `group`: grape variety system (Kyoho-type / Shine Muscat-type)
- `grain_diameter_mm`: measured grain diameter (mm)
- `bunch_length_cm`: bunch length (cm)
- `bunch_width_cm`: bunch width (cm)
- `year_introduced`: year the variety was introduced
- `price_yen_with_tax`: retail price (JPY, tax included)

### Variety System Classification

Grape varieties were classified into two major variety systems
(Kyoho-type and Shine Muscat-type) based on known breeding lineages
and market classification.

The following rule-based approach was used to assign each variety
to a variety system.

```python
# List of varieties classified as Shine Muscat-type (Green Group)
GREEN = "Green Group"
PURPLE = "Purple Group"

green_list = [
    'Queen Seven', 'Haneou', 'Queen Rouge', 'Shine Muscat',
    'Fujinokagayaki', 'Seto Giants', 'Shigyoku', 'Shinku', 'Kaiji',
    'Queen Roug', 'Scarlet', 'Red Shine Muscat', 'Suiho',
    'Sunshine Red', 'Ving', 'Wasekaiji', 'Pizzutello Bianco',
    'Nouvelle Rose', 'Muscat Noir', 'Miwahime'
]

def grape_group(variety):
    if variety in green_list:
        return GREEN
    else:
        return PURPLE

# Apply classification
df["group"] = df["grape_variety"].apply(grape_group)
```

Varieties not included in the predefined list were treated as Kyoho-type (Purple Group),
